1.Crie um dicionário com informações de um aluno (nome, idade e curso). Em seguida, adicione o campo 'nota' ao dicionário e print o resultado.

```python
aluno = {
    "nome": "Pedro",
    "idade": 25,
    "curso": "Ingles"
}
aluno["nota"] = 8.5

print(aluno)
```

2.Crie um dicionário com 3 produtos e seus preços. Mostre o preço do produto 1.

```python
produtos=[{"nome":"Abacate", "preco":15.50},
          {"nome":"Pera", "preco":12.20},
          {"nome":"Acem", "preco":150.50}]

print(produtos[1]["preco"])
```
