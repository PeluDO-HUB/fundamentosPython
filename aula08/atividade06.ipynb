2.Crie uma função chamada par_ou_impar que receba um número como parâmetro e retorne "par" se ele for par e "ímpar" se for ímpar.


```python
def  par_ou_impar(num):
    if num%2==0:
        print(f"o numero {num} é par")
    else:
        print(f"o numero {num} é impar")


par_ou_impar(2)
par_ou_impar(1)
```

3.Crie uma função que converta uma temperatura de Fahrenheit para Celsius e para Kelvin e print os dois.


```python
def converter_temp(fah):
    celsius = (fah - 32) * 5/9
    kelvin = celsius + 273.15
    
    print(f"Celsius: {celsius:.2f}°C")
    print(f"Kelvin: {kelvin:.2f}°K")

converter_temp(70)
converter_temp(65.3)
```
