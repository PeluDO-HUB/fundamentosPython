1.Crie uma função contar_pares(*n) que conte e retorne quantos números pares existem nos parâmetros passados.

```python
def  contar_pares(*num):
    contadorPar=0
    contadorImpar=0
    for num in num:
        if num%2==0:
            contadorPar+=1
        else:
            contadorImpar+=1
    
    print(f"Numeros pares: {contadorPar}\nNumeros Impares: {contadorImpar}")


contar_pares(1,2,3,4,6,8)
```

2.Crie uma função cubica(*nums) que devolve uma lista com o cubo de cada número recebido. Use for.


```python
def cubica(*nums):
    lista=[]
    for num in nums:
        lista.append(num**3)
    print(lista)

cubica(1,2,3,4,5,6)
```

3.Crie uma função lista_compras(**produtosValor) que receba uma lista de compras (produtos e valor) e imprima-os de forma formatada junto com o valorr total da compra.


```python
def lista_compras(**produtosValor):
    total=0
    for produto, valor in produtosValor.items():
        print(f"{produto}: R$ {valor:.2f}")
        total+=valor
    print(f"Valor total da Compra: R$ {total:.2f}")


lista_compras(Banana=5.50,maça=3.03)
```
